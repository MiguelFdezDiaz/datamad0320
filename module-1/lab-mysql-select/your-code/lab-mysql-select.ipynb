```python
import pandas as pd
```

```python
#CHALLENGE 1

df = pd.read_sql_query("""
SELECT au.au_id AS "AUTHOR ID", au.au_lname AS "LAST NAME", au.au_fname 
    AS "FIRST NAME", t.title AS TITLE, p.pub_name AS PUBLISHER 
    FROM titleauthor AS tta
    INNER JOIN authors AS au
    ON au.au_id = tta.au_id
    INNER JOIN titles AS t
    ON t.title_id = tta.title_id
    INNER JOIN publishers AS p
    ON p.pub_id = t.pub_id
""", engine)
df
```

```python
# CHALLENGE 2

df = pd.read_sql_query("""
SELECT au.au_id AS "AUTHOR ID", au.au_lname AS "LAST NAME", au.au_fname 
	AS "FIRST NAME", p.pub_name AS PUBLISHER, count(*)
    FROM titleauthor AS tta
    INNER JOIN authors AS au
    ON au.au_id = tta.au_id
    INNER JOIN titles AS t
    ON t.title_id = tta.title_id
    INNER JOIN publishers AS p
    ON p.pub_id = t.pub_id
	GROUP BY au.au_id, au.au_lname, au.au_fname, p.pub_name 
""", engine)
df
```

```python
# CHALLEGEN 3

df = pd.read_sql_query("""
SELECT au.au_id AS "AUTHOR ID", au.au_lname AS "LAST NAME", au.au_fname 
	AS "FIRST NAME", s.qty AS TOTAL
    FROM titleauthor AS tta
    INNER JOIN authors AS au
    ON au.au_id = tta.au_id
    INNER JOIN titles AS t
    ON t.title_id = tta.title_id
    INNER JOIN sales AS s
    ON t.title_id = s.title_id
	ORDER BY s.qty DESC
    LIMIT 3
""", engine)
df
```

```python
# CHALLENGE BONUS

df = pd.read_sql_query("""
SELECT au.au_id AS "AUTHOR ID", au.au_lname AS "LAST NAME", au.au_fname 
	AS "FIRST NAME", tta.royaltyper
    FROM titleauthor AS tta
    INNER JOIN authors AS au
    ON au.au_id = tta.au_id
    INNER JOIN titles AS t
    ON t.title_id = tta.title_id
    INNER JOIN roysched AS r
    ON r.royalty = t.royalty
    GROUP BY au.au_id, au.au_lname, au.au_fname, tta.royaltyper
    ORDER BY tta.royaltyper DESC
    LIMIT 3
""", engine)
df
```
